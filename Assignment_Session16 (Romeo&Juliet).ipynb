#Assignment3 - Session16 (Romeo&Juliet - list of words in uniform case)

```python
import requests
```

```python
url="https://ia600302.us.archive.org/20/items/romeojuliet01shak/romeojuliet01shak_djvu.txt"
```

```python
respObj=requests.get(url)
```

```python
print(respObj)
```

```python
respData=respObj.text
```

```python
newData1=respData.lower().strip().split('\n')
```

```python
newData2=[]
```

```python
newData3=[]
```

```python
for each in newData1:
    newData2.extend(each.strip().split())
```

```python
for each in newData2:
    newData3.extend(each.strip().strip('.').strip(',').strip('-').strip('—').strip('^').strip(";").strip(")").strip("(").strip("'").split("'s"))
```

```python
print("Total no. of words in Romeo & Juliet:",len(newData3))
```

```python
print("List of words in Romeo & Juliet:",newData3)
```
