#Session15 - check how many times lowercase s appears

```python
#import the libraries requests and json
import requests
```

```python
#enter the website
emailurl="https://dworken.keybase.pub/signedVerifications/emails.txt"
```

```python
#send the request and catch the response
r=requests.get(emailurl)
```

```python
#perform/use response
sitedata=r.text
```

```python
sitedata
```

#Solution 1 - not using count

```python
newdata=sitedata.split('s')
```

```python
print("Solution1 - No. of lowercase s in the emails.txt=",len(newdata)-1)
```

#Solution 2 - convert to list and use count method

```python
list(sitedata)
```

```python
print("Solution2 - No. of lowercase s in the emails.txt=",sitedata.count('s'))
```
